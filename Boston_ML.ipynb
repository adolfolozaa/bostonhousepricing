###Preparing the dataset

```python
dataset = pd.DataFrame(boston.data, columns=boston.feature_names)
```

```python
dataset.head()
```

```python
dataset['Price']=boston.target
dataset.head()
```

```python
dataset.info()
```

```python
#Summarizing the Stats
dataset.describe()
```

```python
#Check the missing values
print(dataset.isnull())
print(dataset.isnull().sum())
```

```python
#Exploratory  data analysis
#Correlation
dataset.corr() 
```

```python
import seaborn as sns
sns.pairplot(dataset)
```

```python
plt.scatter(dataset.CRIM, dataset.Price)
plt.xlabel('Crime Rate')
plt.ylabel('Price')
```

```python
plt.scatter(dataset.RM, dataset.Price)
plt.xlabel('RM')
plt.ylabel('Price')
```

```python
X = dataset.iloc[:,:-1]
y = dataset.iloc[:,-1]
```

```python
X.head()
X.shape
```

```python
y.shape
```

```python
## Train, Test Split
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size =0.3, random_state = 42)
```

```python
X_train
```

```python
## Standarize the dataset
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.fit_transform(X_test)
```

```python
X_train
```

```python
X_test
```

```python
import pickle 
pickle.dump(scaler,open('scaling.pkl', 'wb'))
```

## Model Training ##

```python
from sklearn.linear_model import LinearRegression
```

```python
regression = LinearRegression()
regression.fit(X_train, y_train)
```

```python
## Print the coeficients
print(regression.coef_)
```

```python
print(regression.intercept_)
```

```python
regression.coef_
```

### Revision de conceptos / Inicio

```python
regression.coef_.shape
```

```python
y_calc=np.matmul(X_test, regression.coef_)
print(y_calc)
```

```python
y_calc1=y_calc+ regression.intercept_
print(y_calc1)
```

```python
y_test
```

```python
a=(y_test-y_calc1)/y_test*100
```

```python
np.average(a)
```

### Revision de conceptos / Fin

```python
## on which parameters the model has been trained
regression.get_params()
```

```python
## Prediction with test data
reg_pred = regression.predict(X_test)
```

```python
reg_pred
```

```python
y_test
```

### Asumptions

```python
## Plot a Scatter Plot for prediction
plt.scatter(y_test, reg_pred)
```

```python
## Residuals
residuals = (y_test-reg_pred)
```

```python
residuals
```

```python
## Plot Residuals
sns.displot(residuals, kind='kde')
```

```python
## Scatter plot with respect to prediction and residuals
## uniform Distribution
plt.scatter(reg_pred, residuals)
```

```python
## Performance Metrics

from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error

print(mean_absolute_error(y_test, reg_pred))
print(mean_squared_error(y_test, reg_pred))
print(np.sqrt(mean_squared_error(y_test, reg_pred)))
```

## R square and adjusted R square

```python
from sklearn.metrics import r2_score
score = r2_score(y_test, reg_pred)
print(score)
```

```python
## Display adjusted R-squared
1-(1-score)*(len(y_test)-1)/(len(y_test)-X_test.shape[1]-1)
```
